```python
# Summary of Python Basics and some syntax examples

def section(title):
  print("\n" + "="*len(title))
  print(title)
  print("="*len(title))

# 1. Lists
section("Lists")
fruits = ["apple", "banana", "cherry"]
print("Initial list:", fruits)

# common operations
fruits.append("date")
print("After append:", fruits)
fruits.insert(1, "blueberry")
print("After insert at index 1:", fruits)
print("Slice [1:3]:", fruits[1:3])
print("List comprehension (uppercased):", [f.upper() for f in fruits])

# 2. Tuples
section("Tuples")
point = (10, 20)
print("Tuple (immutable):", point)
# unpacking
x, y = point
print("Unpacked x, y:", x, y)
# single-element tuple
single = (42,)
print("Single-element tuple:", single)

# 3. Dictionaries
section("Dictionaries")
student = {"name": "Alice", "age": 23, "major": "CS"}
print("Dictionary:", student)
print("Access value by key:", student["name"])
student["gpa"] = 3.8
print("After adding key:", student)
print("Iterate keys and values:")
for k, v in student.items():
  print(" ", k, "->", v)

# 4. Sets
section("Sets")
a = {1, 2, 3}
b = {2, 3, 4}
print("Set A:", a)
print("Set B:", b)
print("Union:", a | b)
print("Intersection:", a & b)
print("Difference A - B:", a - b)

# 5. Control flow: if / for / while
section("Control Flow")

n = 7
if n % 2 == 0:
  print(n, "is even")
else:
  print(n, "is odd")

print("For loop over list:")
for idx, fruit in enumerate(fruits):
  print(" ", idx, fruit)

print("While loop (count down):")
count = 3
while count > 0:
  print(" ", count)
  count -= 1
print(" 0 (end)")

# 6. Functions
section("Functions")

def greet(name="Guest"):
  """Return a greeting string."""
  return f"Hello, {name}!"

print(greet("Phat"))
print(greet())

def sum_and_product(a, b):
  """Return multiple values: sum and product."""
  return a + b, a * b

s, p = sum_and_product(3, 5)
print("sum:", s, "product:", p)

# 7. Putting it together: a small example function
section("Mini Project Example: summarize a list of numbers")

def summarize_numbers(nums):
  """
  Given an iterable of numbers, return a dictionary summary:
  - count, sum, min, max, average
  - unique values (as a set)
  """
  nums = list(nums)  # allow any iterable
  if not nums:
    return {"count": 0}
  summary = {
    "count": len(nums),
    "sum": sum(nums),
    "min": min(nums),
    "max": max(nums),
    "average": sum(nums) / len(nums),
    "unique": set(nums)
  }
  return summary

numbers = [1, 2, 2, 3, 4, 4, 4]
print("Numbers:", numbers)
print("Summary:", summarize_numbers(numbers))
```

## Part 2. Object-Oriented Programming (OOP) with Python

```python
# Example of a simple class to represent a network router
# This is a basic implementation to demonstrate object-oriented programming (OOP) concepts in Python.

class Router:
    def __init__(self, name, ip_address):
        self.name = name
        self.ip_address = ip_address
        self.connected_devices = []

    def connect_device(self, device_name):
        self.connected_devices.append(device_name)
        print(f"{device_name} has been connected to {self.name}.")

    def disconnect_device(self, device_name):
        if device_name in self.connected_devices:
            self.connected_devices.remove(device_name)
            print(f"{device_name} has been disconnected from {self.name}.")
        else:
            print(f"{device_name} is not connected to {self.name}.")

    def show_connected_devices(self):
        print(f"Devices connected to {self.name}:")
        for device in self.connected_devices:
            print(f"- {device}")
# Example usage
router1 = Router("Home Router", "192.168.1.1")
router1.connect_device("Laptop")
router1.connect_device("Smartphone")
router1.show_connected_devices()
router1.disconnect_device("Laptop")
router1.show_connected_devices()
```

#### **Exercise 1**: A valid postal code

A valid postal code P have to fullfil both below requirements:

P must be a number in the range from 100000 to 999999  inclusive.
P must not contain more than one alternating repetitive digit pair.
Alternating repetitive digits are digits which repeat immediately after the next digit. In other words, an alternating repetitive digit pair is formed by two equal digits that have just a single digit between them.

Method of implementation, algorithm and approach to solve the problem.
I used a library name regex (usually called "re" for short) to defined the post code based on requirement

```python
regex_integer_in_range = r"[1-9][0-9]{5}$"	# Do not delete 'r'.
regex_alternating_repetitive_digit_pair = r"(\d)(?=\d\1)"	# Do not delete 'r'.

    

import re
P = input()

print (bool(re.match(regex_integer_in_range, P)) 
and len(re.findall(regex_alternating_repetitive_digit_pair, P)) < 2)
```

#### **Exercise 2**: The Minion Game

DescriptionGame Rules

Both players are given the same string, .
Both players have to make substrings using the letters of the string .
Stuart has to make words starting with consonants.
Kevin has to make words starting with vowels.
The game ends when both players have made all possible substrings.

Scoring
A player gets +1 point for each occurrence of the substring in the string .

Method of implementation, algorithm and approach to solve the problem.
Using merely loop to solve this problem

```python
def minion_game(string):
    vowel = 'AEIOU'
    kevin = stuart = 0
    for i in range(len(string)):
        if string[i] in vowel:
            kevin += len(string) - i
        else:
            stuart += len(string) - i
    if kevin > stuart:
        print('Kevin', kevin)
    elif kevin < stuart:
        print('Stuart', stuart)
    else:
        print('Draw')

if __name__ == '__main__':
    s = input()
    minion_game(s)
```
