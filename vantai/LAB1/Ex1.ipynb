## Part 2. Object-Oriented Programming (OOP) with Python

```python
# Example of a simple class to represent a network router
# This is a basic implementation to demonstrate object-oriented programming (OOP) concepts in Python.

class Router:
    def __init__(self, name, ip_address):
        self.name = name
        self.ip_address = ip_address
        self.connected_devices = []

    def connect_device(self, device_name):
        self.connected_devices.append(device_name)
        print(f"{device_name} has been connected to {self.name}.")

    def disconnect_device(self, device_name):
        if device_name in self.connected_devices:
            self.connected_devices.remove(device_name)
            print(f"{device_name} has been disconnected from {self.name}.")
        else:
            print(f"{device_name} is not connected to {self.name}.")

    def show_connected_devices(self):
        print(f"Devices connected to {self.name}:")
        for device in self.connected_devices:
            print(f"- {device}")
# Example usage
router1 = Router("Home Router", "192.168.1.1")
router1.connect_device("Laptop")
router1.connect_device("Smartphone")
router1.show_connected_devices()
router1.disconnect_device("Laptop")
router1.show_connected_devices()
```

#### **Exercise 1**: No Idea
Mô tả đề bài: Cho một mảng số nguyên và hai tập hợp rời nhau $A$ (yêu thích) và $B$ (ghét). Với mỗi phần tử trong mảng, nếu nó thuộc tập $A$ thì mức độ hạnh phúc tăng 1, nếu thuộc tập $B$ thì giảm 1. Mục tiêu là tính tổng mức độ hạnh phúc cuối cùng

Phương pháp giải quyết: * Sử dụng cấu trúc dữ liệu set cho $A$ và $B$ để tối ưu hóa việc kiểm tra sự tồn tại của phần tử (độ phức tạp $O(1)$).
Duyệt qua từng phần tử trong mảng, so khớp với hai tập hợp để cập nhật biến happiness.

```python
if __name__ == '__main__': 
    
    n, m = map(int, input().split())
    arr_n = list(map(int, input().split()))
    arr_A = set(list(map(int, input().split())))
    arr_B = set(list(map(int, input().split())))

    happiness = 0

    for value in arr_n:
        if value in arr_A: 
            happiness += 1
        if value in arr_B: 
            happiness -= 1

    

    print(happiness)
```

#### **Exercise 2**: Write a Function (Leap Year)
Mô tả đề bài: Xác định xem một năm cho trước có phải là năm nhuận theo lịch Gregorian hay không. Điều kiện: Chia hết cho 4 nhưng không chia hết cho 100, trừ khi nó cũng chia hết cho 400.
Phương pháp giải quyết:

Sử dụng các toán tử logic (and, or) và toán tử chia lấy dư (%) để kiểm tra đồng thời các điều kiện: (year % 4 == 0 và year % 100 != 0) hoặc (year % 400 == 0).

Trả về giá trị True nếu thỏa mãn và False nếu ngược lại.

```python
def is_leap(year):
    leap = False
    
    if (year % 4 == 0 and year % 100 != 0) or year % 400 == 0:
        leap = True
    
    return leap

year = int(input())
print(is_leap(year))
```

#### **Exercise 3**: Symmetric Difference
Mô tả đề bài: Cho hai tập hợp số nguyên $M$ và $N$. Tìm "Symmetric Difference" (Hiệu đối xứng) của chúng – tức là những phần tử chỉ xuất hiện ở một trong hai tập hợp nhưng không xuất hiện ở cả hai. Kết quả cần in ra theo thứ tự tăng dần.

Phương pháp giải quyết:Chuyển đổi danh sách đầu vào thành kiểu set.Sử dụng phương thức .difference() để tìm các phần tử có trong $M$ mà không có trong $N$ và ngược lại.Gộp hai kết quả, dùng hàm sorted() để sắp xếp tăng dần và in từng phần tử.

```python
if __name__ == '__main__':
    m = int(input())
    arr_m = list(map(int, input().split()))
    n = int(input())
    arr_n = list(map(int, input().split()))

    set_m, set_n = set(arr_m), set(arr_n)

    result1 = set_m.difference(set_n)
    result2 = set_n.difference(set_m)
    result = sorted(list(result1) + list(result2))
    
    for i in result: 
        print(i)
```

#### **Exercise 4**: Introduction to Sets (Average)
Mô tả đề bài: Tính giá trị trung bình của các chiều cao thực vật khác nhau trong một danh sách. Nếu có các chiều cao trùng lặp, chỉ tính một lần

Phương pháp giải quyết:Sử dụng hàm set() để loại bỏ tất cả các giá trị trùng lặp trong mảng đầu vào, chỉ giữ lại các giá trị duy nhất (distinct heights).Sử dụng hàm sum() để tính tổng và len() để lấy số lượng phần tử của tập hợp đó.Tính trung bình bằng công thức: $Average = \frac{Sum}{Total\ Elements}$ và làm tròn đến 3 chữ số thập phân bằng hàm round().

```python
def average(array):
    sum_of_distinct_height = sum(set(arr))
    length = len(set(arr))
    return round(sum_of_distinct_height / length, 3)

if __name__ == '__main__':
    n = int(input())
    arr = list(map(int, input().split()))
    result = average(arr)
    print(result)
```
